```python
import sqlite3
import pandas as pd

# CSV file path
csv_file = 'Companies.csv'

# SQLite database connection
conn = sqlite3.connect('database.db')

# Read CSV into a pandas DataFrame
df = pd.read_csv(csv_file)

# Write the DataFrame to the SQLite database
df.to_sql('US_companies', conn, if_exists='replace', index=False)

# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('database.db')  # Replace 'your_database.db' with your actual database file

# Create a cursor object
cursor = conn.cursor()

# Execute a query to fetch data from the database
query = "SELECT * FROM US_companies"  # Replace 'your_table' with the name of your table
cursor.execute(query)

# Fetch all the rows
rows = cursor.fetchall()

# Get the column names from the cursor description
columns = [col[0] for col in cursor.description]

# Create a DataFrame using the fetched data and column names
df = pd.DataFrame(rows, columns=columns)

# Save the DataFrame to a CSV file
df.to_csv('output.csv', index=False)  # Replace 'output.csv' with the desired output file name

# Close the database connection
conn.close()

# Display the DataFrame
df
```

```python
from flask import Flask, render_template
import sqlite3
app = Flask(__name__)
 
def dbConnection():
    conn = sqlite3.connect('database.db')
    conn.row_factory = sqlite3.Row
    return conn

@app.route("/")
def index():
    conn = dbConnection()
    query = f'SELECT * FROM US_companies'
    data = conn.execute(query).fetchall()
 
    return render_template('index.html', data = data)

if __name__ == '__main__':
    app.run(debug = True)
```
